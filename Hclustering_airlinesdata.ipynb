### Read data set

```python
airlines=pd.read_csv('EastWestAirlines.csv')
airlines
```

```python
airlines.info()
```

```python
### Dropping ID# column
airline=airlines.drop(['ID#'],axis=1)
airline
```

### Normalization

```python
from sklearn.preprocessing import normalize
airline_norm=pd.DataFrame(normalize(airline),columns=airline.columns)
airline_norm
```

### DENDROGRAM

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch  
z = linkage(airline_norm, method="average",metric="euclidean")
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
z,
)
plt.show()     
```
